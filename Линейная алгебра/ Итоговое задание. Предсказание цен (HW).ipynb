```python
import pandas as pd
import numpy as np

# Определяем данные для расчетов

X = np.array([
    [3, 51, 3, 0, 1, 0],
    [1, 30, 1, 0, 1, 0],
    [2, 45, 2, 0, 1, 0],
    [3, 55, 1, 0, 1, 0],
    [1, 45, 3, 1, 0, 0],
    [3, 100, 3, 1, 0, 0],
    [2, 71, 2, 1, 0, 0],
    [1, 31, 2, 0, 0, 1],
    [3, 53, 5, 0, 0, 1],
    [1, 33, 3, 0, 0, 1],
    [2, 43, 5, 0, 0, 1]
])
X_Cols = ['Комнаты',	'Площадь',	'Этаж',	'Центр',	'Спальный1',	'Спальный2']

Y = np.array([2200, 1600, 1900, 2000, 4500, 7000, 5000, 1700, 2100, 1500, 2000])
```

# Задание 1

```python
#ваш код

def multiplication_order(A, B, C):
    '''
    :param A: Матрица A
    :param B: Матрица B
    :param C: Матрица C
    :return: Возвращает строку "(AxB)xC", 
    если количество умножений элементов матриц при умножении 
    меньше либо равно количеству умножений, если выполнять их в порядке. 
    В противном случае верните строку "Ax(BxC)".
    '''
    # полчаем размерности матриц
    a_r, a_c = A.shape
    b_r, b_c = B.shape
    c_r, c_c = C.shape
    # расчитываем количество элементарных операций при разных порядках умножения
    var_1 = a_r * a_c * b_c + a_r * b_c * c_c   
    var_2 = b_r * b_c * c_c + a_r * b_c * c_c
    # выбораем вариант умножения с наименьшим количеством операций
    return "(AxB)xC" if  var_1 <= var_2 else "Ax(BxC)"

# Пример 

A = np.array([[1, 2]])
B = np.array([[2], [1]])
C = np.array([[5]])

multiplication_order(A, B, C)
```

# Задание 2


```python
#ваш код

def best_worst(X, Y):
    '''
    :param X: матрица признаков
    :param Y: вектор цен
    :return: Кортеж состоящий из
    max_corr_idx - номер признака наиболее коррелирующего с ценой
    min_corr_idx - номер признака наименее коррелирующего с ценой по абсолютному значению
    '''
    # обьединяем данные в единый датафрейм
    df = pd.DataFrame.from_records(X, columns=X_Cols)
    df['Цена'] = Y.tolist()
    # расчитываем матрицу корреляций
    corr = df.corr()
    # визуализируем данные для наглядности
    display(df)
    display(corr.style.background_gradient(cmap='coolwarm'))
    # удаляем столбец с ценой после расчета корреляций
    corr = corr.drop('Цена', axis=1)
    # получаем последнюю строку фрейма 
    # содержащую коэффициенты корреляций признаков по отношению к цене
    # и переводим величины в абсолютные значения
    price_corr = np.abs(corr.values[-1:])    
    # возвращаем индексы параметров с предельными значениями
    return np.argmax(price_corr), np.argmin(price_corr)

# Пример 

max_corr_idx, min_corr_idx = best_worst(X, Y)

print(f'''
max_corr_idx: {max_corr_idx}
min_corr_idx: {min_corr_idx}
''')
```

```python
#ваш код

def corr_rank(X): 
    '''
    :param X: Матрица признаков
    :return:  Ранг корреляционной матрицы
    '''
    # расчитываем корреляционную матрицу 
    corr = np.corrcoef(X, rowvar=False)
    # возращаем ранг полученой матрицы
    return np.linalg.matrix_rank(corr)

# Пример 
corr_rank(X)
```

# Задание 4

```python
#ваш код

def inverse_matrix(A):
    '''
    :param A: матрица в виде numpy-массива.
    :return:  возращает None, если матрица необратима, либо обратную матрицу в виде numpy-массива
    '''
    # Проверяем критерий обратимости и возращаем обратную матрицу если возможно
    return None if abs(np.linalg.det(A)) < 0.001 else np.linalg.inv(A)

# Пример 
A = np.array([
    [1, 2], 
    [2, 1]
])
print(f'''
Исходная матрица:
{A}

Обратная матрица:
{inverse_matrix(A)}
''')
```

# Задание 5

```python
#ваш код

def fit_model(X, Y):
    '''
    :param X: матрица признаков
    :param Y: вектор цен
    :return: возращает numpy-массив с оптимальными коэффициентами , найденными методом OLS
    '''
    return inverse_matrix(X.T @ X) @ X.T @ Y

# Пример 
print(fit_model(X, Y))
```
